Задание 7 (1 балл): придумать новый признак (колонку) для наших данных.

Еще один возможный признак, который можно добавить в данные - это время суток, когда была совершена покупка. Для этого, можно использовать информацию о времени покупки, которая уже есть в наших данных. Например, можно разделить сутки на несколько временных интервалов (например, утро, день, вечер, ночь) и присвоить каждой покупке соответствующий интервал.

```python
data['Time'] = pd.to_datetime(data['Time'])

## здесь я вычислила время суток
bins = [0, 6, 12, 18, 24]
labels = ['ночь', 'утро', 'день', 'вечер']
data['время суток'] = pd.cut(data['Time'].dt.hour, bins=bins, labels=labels, include_lowest=True)

## тут вывела первые 5 строк таблицы с новым признаком
data.head(5)
```

В моем случае, со временем суток, это может помочь определить, в какое время суток наиболее популярны различные товары и категории товаров, что может быть полезно при планировании распределения товаров в магазине и управлении запасами

Задание 12 (1 балл): создать матрицу 10x10, в которой все значения нули, кроме границ. Граничные элементы пускай будут равны единицам.

Задание 14 (0.25 балла): создать массив из 50 элементов (заполнить случайными целыми числами от 0 до 10, 10 не включаем) и для него найти среднее, стандартное отклонение и медиану.

```python
# Создала массив из 50 случайных целых чисел от 0 до 9
a = np.random.randint(0, 10, 50)

# Нашла среднее значение
k = np.mean(a)

# Нашла стандартное отклонение
s = np.std(a)

# Нашла медиану
med = np.median(a)

print("Массив:", a)
print("Среднее значение:", k)
print("Стандартное отклонение:", s)
print("Медиана:", med)
```
