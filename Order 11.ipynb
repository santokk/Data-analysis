### Я полностью обработала наш dataframe, теперь начнем проверять гипотезы:

# Гипотеза №1
Есть взаимосвязь между расстоянием от старой Москвы и тем оформлен ли объект недвижимости. Чем больше расстояние, тем выше шансы, что объект не оформлен. 

+ Зависимая переменная: оформлен/не оформлен
+ Предиктор: расстояние от старой Москвы


```python
df.head()
```

```python
df['Снят/необходимо снять с контроля'].unique()
```

```python
# рассмотрим теперь только те данные у которых нет проблематики (удаляю те у которых 'Снят/необходимо ...' == 'Снято с контроля','Некапитальные. Направлено на снятие'; и 'Поселение\n(без Москвы)' == 'Москва')

data_filtered = df[~(
    (df['Снят/необходимо снять с контроля'].isin(['Снято с контроля', 'Некапитальные. Направлено на снятие'])) | 
    (df['Поселение\n(без Москвы)'] == 'Москва')
)]

data_filtered.shape
```

```python
data_filtered['Кад. номер ОКС'].unique()
```

### Сначала построю бинарную регрессию

```python
# создам бинарную переменную оформлен/не оформлен:
data_filtered['Статус оформления'] = ((data_filtered['Кад. номер ЗУ'] == '+') & (data_filtered['Кад. номер ОКС'] == '+')).astype(int)
data_filtered.head()
```

```python
# посомтрим сколько их чтобы знать предварительно насколько хороша будет работать бинарная регрессия
oformlen_counts = data_filtered['Статус оформления'].value_counts()
print(oformlen_counts)
```

Видим, что оформленных объектов больше чем неоформленных на 13594. Исходя из чего можно сказать, что наша будущая бинарная регрессия будет недостаточно точна

```python
# изменим тип колонки для бинарной регрессии
data_filtered['Удаленность от Москвы (км). Точка отсчета центр Москвы (крадчайший путь)'] = pd.to_numeric(data_filtered['Удаленность от Москвы (км). Точка отсчета центр Москвы (крадчайший путь)'], errors='coerce')
data_filtered.info()
```

```python
# делим на обучающую и тестовые выборки:
from sklearn.model_selection import train_test_split

X = data_filtered[['Удаленность от Москвы (км). Точка отсчета центр Москвы (крадчайший путь)']]
y = data_filtered['Статус оформления']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression

model = LogisticRegression()
model.fit(X_train, y_train)
predictions = model.predict(X_test)
```

```python
from sklearn.metrics import classification_report

print(classification_report(y_test, predictions))
```

По данным результатам бинарной регрессии можно сказать, что модель очень хорошо справляется с предсказанием класса 0 (Не оформлен) (с точностью 92% и полнотой 100%), но совершенно не определяет класс 1 (Оформлен) (с точностью и полнотой 0%). Это может указывать на несколько возможных проблем или особенностей данных:

1. **Несбалансированность классов:** данные сильно несбалансированными, что видно по разнице в количестве наблюдений для двух классов (4377 против 356). Модели машинного обучения часто испытывают трудности с обучением на несбалансированных данных, так как они склонны предсказывать более многочисленный класс

2. **Недостаточная дискриминационная способность признаков:** используемые признаки (в данном случае — расстояние от старой Москвы) не обладают достаточной дискриминационной способностью для различения двух классов эффективно

Что касается нашей гипотезы:

1. **Валидация гипотезы:** Гипотеза утверждала, что существует взаимосвязь между расстоянием от старой Москвы и оформленностью объекта недвижимости — чем больше расстояние, тем выше шансы, что объект не оформлен. Однако результаты модели показывают, что модель (бинарная регрессия) не смогла адекватно классифицировать объекты как неоформленные (класс 1), поскольку для этого класса получены нулевые значения precision и recall.

2. **Ограниченность данных:** Высокая точность (accuracy) модели в основном обусловлена способностью корректно классифицировать оформленные объекты (класс 0), а не способностью различать оформленные и неоформленные объекты. Это указывает на то, что модель в основном предсказывает наиболее частый класс (оформленные объекты) и не выявляет достаточно информации для классификации неоформленных объектов.

3. **Вывод по гипотезе:** Исходя из данных, нельзя утверждать, что гипотеза подтверждена, поскольку модель не демонстрирует эффективности в предсказании неоформленности объектов на основе расстояния от Москвы

### Раз наша бинарная регрессия не смогла ничего ТОЧНО сказать о нашей гипотезе, напишем тогда мультиномиальную регрессию:

```python
data_filtered['Статус населенного пункта (СНТ/ОНТ/ДНТ или город/деревня)'].unique()
```

Для построения мультиноминальной регрессии, я сначала создала новую переменную для статуса оформления, которая будет иметь три категории в соответствии с описанными условиями в гипотезе. Затем я использую эту переменную как зависимую в мультиноминальной регрессионной модели. 
Вот как я закодирую данную новую переменную:

+ Оформлен ЗУ и не оформлен ОКС (кодируем как 0)
+ Не оформлен ЗУ, оформлен ОКС (кодируем как 1)
+ Ничего не оформлено (кодируем как 2)

```python
def categorize_status(row):
    if row['Кад. номер ЗУ'] == '+' and row['Кад. номер ОКС'] == '-':
        return 0  # Оформлен ЗУ и не оформлен ОКС
    elif row['Кад. номер ЗУ'] == '-' and row['Кад. номер ОКС'] == '+':
        return 1  # Не оформлен ЗУ, оформлен ОКС
    return 2  # Ничего не оформлено

data_filtered['Мультиноминальный статус'] = data_filtered.apply(categorize_status, axis=1)
```

```python
# посомтрим сколько их чтобы знать предварительно насколько хороша будет работать бинарная регрессия
oformlen_counts = data_filtered['Мультиноминальный статус'].value_counts()
print(oformlen_counts)
```

```python
# Посмотрим как выглядит наш датасет:
data_filtered.head()
```

```python
# теперь строим мультиноминальную регрессию:
from sklearn.linear_model import LogisticRegression

X = data_filtered[['Удаленность от Москвы (км). Точка отсчета центр Москвы (крадчайший путь)']]
y = data_filtered['Мультиноминальный статус']

model = LogisticRegression(multi_class='multinomial', solver='lbfgs')
model.fit(X, y)
predictions = model.predict(X)
```

```python
# отчет о классификации:
from sklearn.metrics import classification_report
print(classification_report(y, predictions))
```

```python
# коэффиценты регрессии:
list(model.coef_)
```

```python
# матрица ошибок:
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y, predictions)
print(cm)
```

   - Модель имеет хорошую точность (precision) и полноту (recall) только для класса 0, что указывает на то, что она успешно идентифицирует случаи, когда ЗУ оформлен, а ОКС не оформлен.
   - Классы 1 и 2 (не оформлен ЗУ, оформлен ОКС и ничего не оформлено, соответственно) не идентифицируются вообще. Модель не смогла правильно классифицировать ни одного случая для этих классов, что отражено в нулевых значениях precision и recall.

2. **Коэффициенты регрессии:**
   - Коэффициент для класса 0 положительный, что указывает на положительное влияние расстояния на вероятность оформления ЗУ без оформления ОКС.

   - Коэффициент для класса 2 близок к коэффициенту класса 0, но, поскольку модель не смогла правильно идентифицировать случаи для класса 2, трудно сделать определенные выводы по этому коэффициенту.

3. **Матрица ошибок:**
   - Вся диагональ матрицы ошибок за исключением первой ячейки заполнена нулями, что говорит о том, что модель предсказывает только класс 0 и игнорирует классы 1 и 2.

Выводы:
- Модель эффективно распознает только объекты из класса 0 (оформлен ЗУ, не оформлен ОКС) и не распознает объекты из классов 1 и 2. Это может свидетельствовать о нехватке информативности предиктора (расстояния от Москвы) для различения этих специфических статусов оформления или о необходимости дополнительного анализа и возможно применения других методов или дополнительных предикторов для более эффективной классификации.

```python
# для начала добавим характерный столбец:
data_filtered['Статус СНТ/ОНТ'] = data_filtered['Статус населенного пункта (СНТ/ОНТ/ДНТ или город/деревня)'].apply(lambda x: 1 if x in ['СНТ', 'ОНТ', 'деревня'] else 0)

oformlen_counts = data_filtered['Статус СНТ/ОНТ'].value_counts()
print(oformlen_counts)
```
